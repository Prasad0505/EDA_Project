**Droping The unnacessary Columns**

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

**Droping the duplicates**

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
from markupsafe import escape
def myfun(value):
    if(value=='NEW' or value=='-'):
       return np.nan
    else:
       value=str(value).split('/')
       value=value[0]
       return float(value)

df['rate']=df['rate'].apply(myfun)
df['rate'].head()
```

**Filling Null Values in Rate Column With Mean**

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

**Dropping Null Values**

```python
df.dropna(inplace=True)
df.head()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost_of_Two_people','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df.drop('listed_in(city)',inplace=True,axis=1)
```

```python
df['cost_of_Two_people'].unique()
```

```python
def costfun(value):
          value=str(value).replace(',','')
          return float(value)

df['cost_of_Two_people']=df['cost_of_Two_people'].apply(costfun)
df.tail()
```

```python
df['rate'].unique()
```

```python
df['cost_of_Two_people'].unique()
```

**Cleaning Rest Type Column**

```python
df['rest_type'].value_counts()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

**Making Rest Type less than 1000 in frequency as others**

```python
def handel_rest_type(value):
   if value in rest_type_lessthan1000:
      return 'others'
   else:
      return value

df['rest_type']=df['rest_type'].apply(handel_rest_type)
df['rest_type'].value_counts()
      
```

```python
df.head()
```

**Cleaning Location Column**

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location_lessthan300=location[location<300]
location_lessthan300
```

**Making location less than 300 in frequency as others**

```python
def handel_location(value):
       if value in location_lessthan300:
             return 'others'
       else:
             return value

df['location']=df['location'].apply(handel_location)
df['location'].value_counts()
```

**Cleaning Cuisines Column**

```python
df['cuisines'].value_counts()
```

```python
cuisine=df['cuisines'].value_counts()
cuisine_lessthan_100=cuisine[cuisine<100]

def handle_cuisine(value):
    if value in cuisine_lessthan_100:
                    return 'others'
    else:
                return value

df['cuisines']=df['cuisines'].apply(handle_cuisine)
df['cuisines'].value_counts()
```
